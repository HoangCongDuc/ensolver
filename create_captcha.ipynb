```python
from captcha import ImageCaptcha
from PIL import ImageDraw
```

```python
generator = ImageCaptcha(width=400, height=150, font_sizes=(70, 80, 90))
```

```python
img, bboxes = generator.generate_image('Ab2T4', return_bbox=True)
```

```python
img
```

```python
img_box = img.copy()
draw = ImageDraw.Draw(img_box)
for x1, y1, w, h in bboxes:
    bbox = (x1, y1, x1 + w, y1 + h)
    draw.rectangle(bbox, outline='red')

img_box
```
